```python
import pandas as pd
import plotly.express as px 
```

```python
spotify_data = pd.read_csv("../spotify_tracks.csv")
```

```python
print(spotify_data.head())
```

```python
hist_fig = px.histogram(spotify_data, 
                        x = "popularity",
                        nbins=100,
                        log_y=True,
                        color_discrete_sequence=['#1DB954'])
hist_fig.update_layout(title= "Popularity among more than 100,000 songs on Spotify ",
                  plot_bgcolor="#121212",
                  paper_bgcolor="#121212",
                  font_color="white",
                  title_font_size= 20,
)
hist_fig.update_xaxes(gridcolor="#333333")
hist_fig.update_yaxes(gridcolor="#333333")
hist_fig.show()
```

```python
matrix_fig= px.scatter_matrix(spotify_data.sample(1000, random_state= 30),
                               dimensions=["danceability", "loudness", "valence", "energy"],
                               title="<b>Loudness and intensity move together - mood floats free</b><br><sub>Audio features across 1,000 sampled tracks · Spotify dataset</sub>",
                               labels={'danceability': 'Danceability',
                                       'loudness':'Loud(dB)',
                                       'valence':'Mood',
                                       'energy':'Intensity'},
                                       opacity=0.4,
                                       template="plotly_dark",
                                       color_discrete_sequence=['#1DB954']
                                       
                                       )
matrix_fig.update_layout(
    width=1100,
    height=700, 
    )
matrix_fig.update_xaxes(gridcolor="#222222")
matrix_fig.update_yaxes(gridcolor="#222222")
matrix_fig.show()
```

```python
scatter_fig = px.scatter(spotify_data.sample(1000, random_state= 30),
                         x="energy",
                         y="valence",
                         labels={'energy':'Intensity',
                                 'valence':'Mood',
                                 'loudness':'Loud(dB)'},
                         color="loudness",
                         hover_name="track_name",
                         hover_data=["artists", "album_name"],
                         opacity=0.4,
                         title="<b> The mood of a song is more related to its energy than its loudness</b><br><sub>Energy and mood across 1,000 sampled tracks · Spotify dataset</sub>",
                         template="plotly_dark",
                         color_continuous_scale="Viridis"
)
scatter_fig.update_xaxes(gridcolor="#222222")
scatter_fig.update_yaxes(gridcolor="#222222")
scatter_fig.show()
```
